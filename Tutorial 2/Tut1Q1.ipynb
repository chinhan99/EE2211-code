```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy import stats

data = pd.read_csv("GovernmentExpenditureonEducation.csv")
slope, intercept, r_value, p_value, std_err = stats.linregress(data.year, data.total_expenditure_on_education)

#print("Contents in csv file:", data)
```

```python
plt.scatter(data.year, data.total_expenditure_on_education, label ="Education expenditure")
plt.xlabel('X')
plt.ylabel('Y')

plt.plot(data.year, slope*data.year + intercept, color='red', label='Regression Line')
plt.legend()
plt.show()
```
