```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
cars = ["Omnibuses", "Excursion buses", "Private buses"]

data = pd.read_csv("AnnualMotorVehiclePopulationbyVehicleType.csv")
#pandas change csv file into a dataframe for easy manipulation of data .

df = pd.DataFrame(data)
filtered_data = data[data["type"].isin (cars)]


"""
pivot vs pivot_table method both works. but pivot_table allows aggregation.
"""
#pivot_data = filtered_data.pivot_table(index = "year", columns = "type", values= "number", aggfunc ='sum')
pivot_data = filtered_data.pivot(index = "year", columns = "type", values= "number")

#https://builtin.com/data-science/pandas-pivot-tables
#https://medium.com/@filip.sekan/pivot-vs-pivot-table-methods-in-pandas-definitions-examples-differences-dcc9e14e667f#:~:text=pivot%20is%20a%20simpler%20method,values%2C%20whereas%20pivot%20does%20not.
#print (pivot_data)

plt.figure(figsize=(10, 6))
ax = plt.gca()

pivot_data.plot(kind='line', ax=ax, marker='o')
# Add labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Number of Cars')
ax.set_title('Number of Cars by Car Model Over Years')

# Add legend
ax.legend(title='Car Model')

# Display the plot
plt.tight_layout()
plt.show()
```
